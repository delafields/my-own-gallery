```python
with open('image_links_old.json') as json_file:
    old_data = json.load(json_file)

for palette_key in old_data.keys():

    with open('image_links.json') as infile:
        processed_palettes = json.load(infile)

    if palette_key in processed_palettes.keys():
        print("dupe")
        continue

    broken_up = [palette_key[i:i+6] for i in range(0, len(palette_key), 6)]
    palette_url = "-".join(broken_up)
    url = "https://artsexperiments.withgoogle.com/artpalette/colors/" + palette_url

    print("getting images at: ", url)
    
    hex_codes = ["#" + hex_code for hex_code in broken_up]

    # go to this palette's page, scroll down
    driver.get(url)
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, window.scrollY + 10000)")
    time.sleep(6)

    # get image detail pages
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    current_palette_images = []
    image_pages = soup.find_all("a", {"class": "more-details"})

    # go to specific image pages
    for image_page in image_pages:
        current_image = {}
        # get image title, remove non-ascii
        encoded_title = image_page["title"].encode('ascii', 'ignore')
        title = encoded_title.decode()
        current_image["title"] = title

        driver.get(image_page["href"])
        time.sleep(1)

        # get the image in the noscript tag
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        noscript = soup.find('noscript')

        # check if this image exists
        if noscript and noscript.find("img"):
            image_source = noscript.find("img")["src"]
            if "http" not in image_source:
                image_source = "http:" + image_source
            print(image_source)
            current_image["link"] = image_source
            current_palette_images.append(current_image)
        # if no image, continue
        else:
            continue
    print("DONE WITH IMAGES")
    
    # add new palette data to json file
    with open("image_links.json") as infile:
        palettes = json.loads(infile.read())
        palettes[palette_key] = {"palette_hex_codes": hex_codes,
                                 "images": current_palette_images}
    with open("image_links.json", "w") as outfile:
        json.dump(palettes, outfile)
```
